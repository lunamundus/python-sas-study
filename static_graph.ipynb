# 정규 분포 그래프

## 정규 분포(Normal Distribution)

```python
import matplotlib.pyplot as plt
import numpy as np

from scipy.stats import norm

x = np.arange(0, 20, 0.01)

plt.figure(figsize=(12, 8))
plt.xlabel("x")
plt.ylabel("f(x)")

plt.xticks(np.arange(0, 21, 2), ['', '', '', '', '', 'µ', '', '', '', '', ''])
plt.yticks(np.arange(0, 0.22, 0.025), ['', '', '', '', '', '', '', '', ''])

plt.rc('axes', labelsize=20)

plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)

plt.ylim(0, 0.2)
plt.xlim(0, 20)

plt.plot(x, norm.pdf(x, loc=10, scale=2))
plt.axvline(10, 0, 1, color='lightgray', linestyle='--', linewidth=2)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

from scipy.stats import norm

x = np.arange(0, 20, 0.01)

plt.figure(figsize=(12, 8))
plt.xlabel("x")
plt.ylabel("f(x)")

plt.xticks(np.arange(0, 21, 2), ['', '', '', '', 'µ₁', '', 'µ₂', '', '', '', ''])
plt.yticks(np.arange(0, 0.22, 0.025), ['', '', '', '', '', '', '', '', ''])

plt.rc('axes', labelsize=20)

plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)

plt.ylim(0, 0.2)
plt.xlim(0, 20)

plt.plot(x, norm.pdf(x, loc=8, scale=2))
plt.plot(x, norm.pdf(x, loc=12, scale=2))

plt.axvline(8, 0, 1, color='lightgray', linestyle='--', linewidth=2)
plt.axvline(12, 0, 1, color='lightgray', linestyle='--', linewidth=2)
plt.show()
```

## 표준 정규 분포(Standard Normal Distribution)

```python
import matplotlib.pyplot as plt
import numpy as np

from scipy.stats import norm

x = np.arange(-4, 4, 0.01)

plt.figure(figsize=(12, 8))
plt.xlabel("x")
plt.ylabel("f(x)")

plt.xticks(np.arange(-4, 5, 1), ['', '', '', '-1', '0', '1', '', '', ''])
plt.yticks(np.arange(0, 0.45, 0.1), ['', '', '', '', ''])

plt.rc('axes', labelsize=20)

plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)

plt.ylim(0, 0.4)
plt.xlim(-4, 4)

plt.plot(x, norm.pdf(x, loc=0, scale=1))
plt.axvline(0, 0, 1, color='lightgray', linestyle='--', linewidth=2)
plt.axvline(-1, 0, 0.582, color='lightgray', linestyle='--', linewidth=2)
plt.axvline(1, 0, 0.582, color='lightgray', linestyle='--', linewidth=2)
plt.show()
```
