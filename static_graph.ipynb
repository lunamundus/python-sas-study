# 정규 분포 그래프

## 정규 분포(Normal Distribution)

```python
import matplotlib.pyplot as plt
import numpy as np

from scipy.stats import norm

x = np.arange(0, 20, 0.01)

plt.figure(figsize=(12, 8))
plt.xlabel("x")
plt.ylabel("f(x)")

plt.xticks(np.arange(0, 21, 2), ['', '', '', '', '', 'µ', '', '', '', '', ''])
plt.yticks(np.arange(0, 0.22, 0.025), ['', '', '', '', '', '', '', '', ''])

plt.rc('axes', labelsize=20)

plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)

plt.ylim(0, 0.2)
plt.xlim(0, 20)

plt.plot(x, norm.pdf(x, loc=10, scale=2))
plt.axvline(10, 0, 1, color='lightgray', linestyle='--', linewidth=2)
plt.show()
```

## 표준 정규 분포(Standard Normal Distribution)

```python
import matplotlib.pyplot as plt
import numpy as np

from scipy.stats import norm

x = np.arange(-4, 4, 0.01)

plt.figure(figsize=(12, 8))
plt.xlabel("x")
plt.ylabel("f(x)")

plt.xticks(np.arange(-4, 5, 1), ['', '', '', '-1', '0', '1', '', '', ''])
plt.yticks(np.arange(0, 0.45, 0.1), ['', '', '', '', ''])

plt.rc('axes', labelsize=20)

plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)

plt.ylim(0, 0.4)
plt.xlim(-4, 4)

plt.plot(x, norm.pdf(x, loc=0, scale=1))
plt.axvline(0, 0, 1, color='lightgray', linestyle='--', linewidth=2)
plt.axvline(-1, 0, 0.582, color='lightgray', linestyle='--', linewidth=2)
plt.axvline(1, 0, 0.582, color='lightgray', linestyle='--', linewidth=2)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

from scipy.stats import norm

x = np.arange(-4, 4, 0.01)
y = norm.pdf(x, loc=0, scale=1)

plt.figure(figsize=(12, 8))
# plt.xlabel("x")
# plt.ylabel("f(x)")

plt.xticks(np.arange(-4, 5, 1), ['', '', r'$-Z_{\alpha/2}$', '', '0', '', r'$Z_{\alpha/2}$', '', ''])
plt.yticks(np.arange(0, 0.45, 0.1), ['', '', '', '', ''])

plt.fill_between(x, y, where=(x>=2)|(x<=-2), alpha=0.2)
plt.fill_between(x, y, where=(-2<x)&(x<2), alpha=0.2, color='green')

plt.rc('axes', labelsize=20)

plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
plt.gca().spines['left'].set_visible(False)

plt.gca().axes.yaxis.set_visible(False)

plt.ylim(0, 0.4)
plt.xlim(-4, 4)

plt.plot(x, norm.pdf(x, loc=0, scale=1))
plt.axvline(0, 0, 1, color='lightgray', linestyle='--', linewidth=2)
plt.axvline(-2, 0, 0.130, color='lightgray', linestyle='--', linewidth=2)
plt.axvline(2, 0, 0.130, color='lightgray', linestyle='--', linewidth=2)
plt.show()
```

# 표본 분포

## 카이 제곱 분포(Chi-square Distribution)

```python
import numpy as np

import scipy.stats as stats
import matplotlib.pyplot as plt

db = np.linspace(0.5, 50, 100)

y = stats.chi2(10).pdf(db)

plt.figure(figsize=(12, 8))
plt.plot(db, y)

plt.xticks(np.arange(0, 51, 10), ['0', '', '', '', '', ''])
plt.yticks(np.arange(0, 0.11, 0.02), ['0', '', '', '', '', ''])

plt.xlim(0, 50)
plt.ylim(0, 0.1)

plt.xlabel(r'$\chi^2$', loc='right')
plt.ylabel(r'$f(\chi^2)$', loc='center', fontdict={'rotation': 0}, labelpad=20)

plt.rc('axes', labelsize=20)

plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)

plt.show()
```

```python
import numpy as np

import scipy.stats as stats
import matplotlib.pyplot as plt

db = np.linspace(0.5, 50, 100)

dfs = [1, 5, 10, 20, 30]

plt.figure(figsize=(12, 8))

for df in dfs:
    plt.plot(db, stats.chi2(df).pdf(db), label=r'$\chi^2$(' + str(df) + ')')

plt.legend()

plt.xticks(np.arange(0, 51, 10), ['0', '', '', '', '', ''])
plt.yticks(np.arange(0, 0.41, 0.1), ['0', '', '', '', ''])

plt.xlim(0, 50)
plt.ylim(0, 0.4)

plt.xlabel(r'$\chi^2$', loc='right')
plt.ylabel(r'$f(\chi^2)$', loc='center', fontdict={'rotation': 0}, labelpad=20)

plt.rc('axes', labelsize=20)

plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)

plt.show()
```

## t 분포(t Distribution, Student-t Distribution)

```python
import numpy as np

import scipy.stats as stats
import matplotlib.pyplot as plt

db = np.linspace(-5, 5, 101)

y = stats.t(3).pdf(db)

plt.figure(figsize=(12, 8))

plt.plot(db, y)

plt.xticks(np.arange(-6, 7, 2), ['', '', '', '0', '', '', ''])
plt.yticks(np.arange(0, 0.5, 0.1), ['0', '', '', '', ''])

plt.xlim(-5, 5)
plt.ylim(0, 0.4)

plt.xlabel(r't', loc='right')
plt.ylabel(r'$f(t)$', loc='center', fontdict={'rotation': 0}, labelpad=20)

plt.rc('axes', labelsize=20)

plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)

plt.show()
```

```python
import numpy as np

import scipy.stats as stats
import matplotlib.pyplot as plt

db = np.linspace(-5, 5, 101)

y = stats.t(3).pdf(db)

plt.figure(figsize=(12, 8))

plt.plot(db, y, label='t Distribution')
plt.plot(db, stats.norm.pdf(db, loc=0, scale=1), label='Normal Distribution')

plt.legend()

plt.xticks(np.arange(-6, 7, 2), ['', '', '', '0', '', '', ''])
plt.yticks(np.arange(0, 0.5, 0.1), ['0', '', '', '', ''])

plt.xlim(-5, 5)
plt.ylim(0, 0.4)

plt.xlabel(r't', loc='right')
plt.ylabel(r'$f(t)$', loc='center', fontdict={'rotation': 0}, labelpad=20)

plt.rc('axes', labelsize=20)

plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)

plt.show()
```

## F 분포(F Distribution)

```python
import numpy as np

import scipy.stats as stats
import matplotlib.pyplot as plt

db = np.linspace(0, 5, 201)

n1 = 4
n2 = 50

y = stats.f(n1, n2).pdf(db)

plt.figure(figsize=(12, 8))

plt.plot(db, y)

plt.xticks(np.arange(0, 6, 1), ['0', '', '', '', '', ''])
plt.yticks(np.arange(0, 0.9, 0.1), ['0', '', '', '', '', '', '', '', ''])

plt.xlim(0, 6)
plt.ylim(0, 0.8)

plt.xlabel(r'F', loc='right')
plt.ylabel(r'$f(F)$', loc='center', fontdict={'rotation': 0}, labelpad=20)

plt.rc('axes', labelsize=20)

plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)

plt.show()
```

# 가설 검정

## 기각역, p-value를 이용한 의사 결정

```python
import matplotlib.pyplot as plt
import numpy as np

from scipy.stats import norm

x = np.arange(0, 20, 0.01)

plt.figure(figsize=(12, 8))

plt.xticks(np.arange(0, 21, 2), ['', '', '', '', '', 'µ', '', '', '', '', ''], fontsize=20)
plt.yticks(np.arange(0, 0.22, 0.025), ['', '', '', '', '', '', '', '', ''])

plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
plt.gca().spines['left'].set_visible(False)

plt.gca().axes.yaxis.set_visible(False)

plt.xlim(0, 20)
plt.ylim(0, 0.2)

# 정규분포 그래프
plt.plot(x, norm.pdf(x, loc=10, scale=2))
# plt.text(4, 0.03, 'Rejection\nRegion', horizontalalignment='center', verticalalignment='center', fontsize=15)
plt.annotate("Rejection\nRegion", xy=(5, 0.005),xytext=(2, 0.03), arrowprops=dict(color='black',arrowstyle="-|>"), fontsize=15)
plt.annotate("Rejection\nRegion", xy=(15, 0.005),xytext=(16, 0.03), arrowprops=dict(color='black',arrowstyle="-|>"), fontsize=15)

# 색 + Text
plt.fill_between(x, norm.pdf(x, loc=10, scale=2), where=(x>=14)|(x<=6), alpha=0.7)

# 세로 선 + Text
plt.text(10, 0.05, 'Nonrejection\nRegion', horizontalalignment='center', verticalalignment='center', fontsize=15)

plt.axvline(10, 0, 1, color='lightgray', linestyle='--', linewidth=2)

plt.axvline(6, 0, 0.3, color='lightgray', linestyle='--', linewidth=2)
plt.text(6, 0.07, 'Critical\nValue', horizontalalignment='center', verticalalignment='center', fontsize=15)

plt.axvline(14, 0, 0.3, color='lightgray', linestyle='--', linewidth=2)
plt.text(14, 0.07, 'Critical\nValue', horizontalalignment='center', verticalalignment='center', fontsize=15)

# 그래프 출력
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

from scipy.stats import norm

x = np.arange(-4, 4, 0.01)
y = norm.pdf(x, loc=0, scale=1)

plt.figure(figsize=(12, 8))

plt.xticks(np.arange(-4, 5, 1), ['', '', r'$-Z^{*}$', '', '0', '', r'$Z^{*}$', '', ''], fontsize=20)
plt.yticks(np.arange(0, 0.45, 0.1), ['', '', '', '', ''])

plt.fill_between(x, y, where=(x>=2)|(x<=-2), alpha=0.7)
# plt.fill_between(x, y, where=(-2<x)&(x<2), alpha=0.2, color='green')

plt.annotate("p-value", xy=(-2.5, 0.01),xytext=(-3.5, 0.1), arrowprops=dict(color='black',arrowstyle="-|>"), fontsize=25)
plt.annotate("p-value", xy=(2.5, 0.01),xytext=(-3.5, 0.1), arrowprops=dict(color='black',arrowstyle="-|>"), fontsize=25, bbox=dict(boxstyle="square, pad=0.4", fc="white", ec="black", lw=1))

plt.rc('axes', labelsize=20)

plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
plt.gca().spines['left'].set_visible(False)

plt.gca().axes.yaxis.set_visible(False)

plt.ylim(0, 0.4)
plt.xlim(-4, 4)

plt.plot(x, norm.pdf(x, loc=0, scale=1))
plt.axvline(0, 0, 1, color='lightgray', linestyle='--', linewidth=2)
plt.axvline(-2, 0, 0.130, color='lightgray', linestyle='--', linewidth=2)
plt.axvline(2, 0, 0.130, color='lightgray', linestyle='--', linewidth=2)
plt.show()
```
