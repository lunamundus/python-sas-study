# 그래프 수직선 및 수평선 그리기

## 기본 데이터 생성

```python
import pandas as pd

data = {
    'Name': ['James', 'David', 'Jessica', 'Lisa'],
    'Math' : [40, 70, 90, 60],
    'English' : [100, 50, 80, 60],
}
```

## 수평선 그리기

### 세로 막대그래프 그리기

```python
import matplotlib.pyplot as plt

plt.bar(data['Name'], data['Math'])

plt.title('Math Score')

plt.show()
```

### 수평선 추가하기

```python
import matplotlib.pyplot as plt

average = 65

plt.bar(data['Name'], data['Math'])
plt.axhline(average, xmin=0, xmax=1, color='orange', linestyle='--', linewidth=2)

plt.title('Math Score')

plt.show()
```

```python
import matplotlib.pyplot as plt

average = 65

plt.bar(data['Name'], data['Math'])
plt.hlines(average, xmin=0, xmax=3, color='orange', linestyle='--', linewidth=2)

plt.title('Math Score')

plt.show()
```

## 수직선 그리기

### 가로 막대그래프 그리기

```python
import matplotlib.pyplot as plt

plt.barh(data['Name'], data['English'])

plt.title('English Score')

plt.show()
```

### 수직선 추가하기

```python
import matplotlib.pyplot as plt

average = 72.5

plt.barh(data['Name'], data['English'])
plt.axvline(average, ymin=0, ymax=1, color='orange', linestyle='--', linewidth=2)

plt.title('English Score')

plt.show()
```
